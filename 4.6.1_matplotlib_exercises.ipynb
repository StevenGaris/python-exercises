```python
import matplotlib.pyplot as plt
import math
%matplotlib inline 
```

Use matplotlib to plot the following equation:

y = x^2 âˆ’ x + 2

Add an anotation for the point 0, 0, the origin.

```python
x = list(range(-5,5))
y = [((n**2) - n + 2) for n in x]

plt.scatter(x,y)
plt.plot(x,y,ls=":")

plt.text(-.25, 3 , 'Origin', fontsize=10, color='blue')

plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Functions Subplot')

r = list(range(0,100))
x = [math.sqrt(n) for n in r]
y = [n**3 for n in r]
z = [math.tan(n) for n in r]
s = [(2**n) for n in r]

plt.subplot(221)
plt.plot(r,x,c='r')
plt.title('$y=âˆšx$')

plt.subplot(222)
plt.plot(r,y,c='b')
plt.title('$y=x^3$')

plt.subplot(223)
plt.plot(r,z,c='g')
plt.title('$y=tan(x)$')


plt.subplot(224)
plt.plot(r,s,c='orange')
plt.title('$y = 2^x$')


plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Functions Combined')

r = list(range(0,100))
x = [math.sqrt(n) for n in r]
y = [n**3 for n in r]
z = [math.tan(n) for n in r]
s = [(2**n) for n in r]

plt.xlim(-2.5,25)
plt.ylim(-20,20)
plt.grid(True)
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')


plt.plot(r, x, c='r', label='$y=âˆšx$')
plt.plot(r, y, c='b', label='$y=x^3$')
plt.plot(r, z, c='g', label='$y=tan(x)$')
plt.plot(r, s, c='orange', label='$y = 2^x$')

plt.legend(loc='best')
plt.show()
```
