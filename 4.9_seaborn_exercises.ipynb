```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pydataset import data
import seaborn as sns
```

```python
# Use the iris database to answer the following quesitons:

iris = sns.load_dataset('iris')
iris

```

```python
# What does the distribution of petal lengths look like?

print(sns.distplot(iris.petal_length))
```

```python
# Is there a correlation between petal length and petal width?

iris[['petal_length','petal_width']].corr()

```

```python
# Would it be reasonable to predict species based on sepal width and sepal length? no, low cor

iris[['sepal_length', 'sepal_width']].corr()

```

```python
# Which features would be best used to predict species? sepal len, petal len and width


sns.heatmap(iris.corr(), annot=True, cmap=plt.cm.Greens)
```

```python
# Using the lesson as an example, use seaborn's load_dataset function to load the anscombe data set.
# Use pandas to group the data by the dataset column, and calculate summary statistics for each dataset. 
# What do you notice?

anscombe = sns.load_dataset('anscombe')
anscombe[['dataset', 'x', 'y']].groupby('dataset').describe()
```

```python
# Load the swiss dataset and read it's documentation. 
# Create visualizations to answer the following questions:

# Create a column named is_catholic that holds a boolean value of whether or not the province is Catholic. 
# (Choose a cutoff point for what constitutes catholic)
# Does whether or not a province is Catholic influence fertility? was only 0.4, not relly
# What measure correlates most strongly with fertility? Examination -.064 and Education -.066 both have neg corr

swiss_data = data('swiss')

is_catholic = swiss_data.assign(is_catholic= swiss_data.Catholic > 50)

is_catholic[['is_catholic', 'Fertility']].corr()

sns.heatmap(is_catholic.corr(), annot=True, cmap=plt.cm.Greens)
```

```python
# Using the chipotle dataset from the previous exercise, 
# create a barplot that shows the 4 most popular items and
# the revenue produced by each.

def get_db_url(db, username, hostname, password):
    from sqlalchemy import create_engine
    url = f'mysql+pymysql://{user}:{password}@{host}/{db}'
    return create_engine(url)
```

```python
from env import user, host, password
```

```python
chip_conn = get_db_url('chipotle', user, host, password)
chip_db = pd.read_sql('select * from orders', chip_conn)
chip_db
```

```python
# create a barplot that shows the 4 most popular items and
# the revenue produced by each.

pop_rev = chip_db[['item_name', 'item_price']].groupby('item_name', as_index= False).sum().nlargest(4, 'item_price')

pop_rev


sns.barplot(data= pop_rev, x= 'item_name', y= 'item_price')
```

```python
# Load the sleepstudy data and read it's documentation. 
# Use seaborn to create a line plot of all the individual subject's reaction times and 
# a more prominant line showing the average change in reaction time.

sleepstudy = data('sleepstudy')

avg = sleepstudy[['Days','Reaction']].groupby('Days', as_index= False).mean()
avg_line = sns.lineplot(data=avg, x= 'Days', y= 'Reaction', linewidth= 10)
sns.lineplot(data=sleepstudy, x= 'Days', y= 'Reaction', hue= 'Subject')


```
